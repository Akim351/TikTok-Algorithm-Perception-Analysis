### Handling missing data

```python
tiktok.isnull().sum()
```

```python
tiktok["tiktok_recommendation_accuracy"].fillna(tiktok["tiktok_recommendation_accuracy"].mode()[0], inplace=True)
tiktok["tiktok_repeated_exposure"].fillna(tiktok["tiktok_repeated_exposure"].mode()[0], inplace=True)
tiktok["tiktok_algorithm_influence"].fillna(tiktok["tiktok_algorithm_influence"].mode()[0], inplace=True)
# tiktok["tiktok_content_diversity"].fillna(tiktok["tiktok_content_diversity"].median(), inplace= True)
tiktok["tiktok_algorithm_awareness"].fillna(tiktok["tiktok_algorithm_awareness"].mode()[0], inplace=True)
tiktok["tiktok_algorithm_modification"].fillna(tiktok["tiktok_algorithm_modification"].mode()[0], inplace=True)
tiktok["tiktok_alternative_viewpoints"].fillna(tiktok["tiktok_alternative_viewpoints"].mode()[0], inplace=True)
tiktok["tiktok_recommendation_control"].fillna(tiktok["tiktok_recommendation_control"].mode()[0], inplace=True)



```

```python
# STIData["Height"].fillna(STIData["Height"].median(), inplace=True)
tiktok["Personalized_For You"].fillna(tiktok["Personalized_For You"].median(), inplace=True)
```

```python
# Count occurrences of each category
use_titok= tiktok["Frequency of use"].value_counts()

# Labels and sizes
labels = use_titok.index  # Unique categories
sizes = use_titok.values  # Corresponding counts
colors = ['blue', 'green', 'red', 'purple', 'yellow']  # Adjust or add more colors if needed

# Plot pie chart
plt.figure(figsize=(3, 4))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True)

# Title
plt.title("How often do you use TikTok?")
```
